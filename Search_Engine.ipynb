```python
def preprocess(text):
    text=text.replace("\n"," ").replace("\r"," ")
    text=text.replace("'s"," ")
    punctuationList= '!"#$%&\()*+,-./:;<=>?@[\\]^_{|}~'
    x=str.maketrans(dict.fromkeys(punctuationList," "))
    text=text.translate(x)
    return text
```

```python
from nltk.tokenize import WhitespaceTokenizer
tokenizer_w = WhitespaceTokenizer()
def tokenizeHelper(text):
    text=preprocess(text)
    return tokenizer_w.tokenize(text)

def Tokenize(df, data):
    tokens = pd.DataFrame() 
    for item in data:
        tokens += df[item].apply(tokenizeHelper)
    return tokens

df_tokenized = Tokenize(df, data)
```

```python
df['TokensPlot']=df['Plot'].apply(tokenizeHelper)
df['TokensTitle']=df['Title'].apply(tokenizeHelper)
df['TokensOrigin']=df['Origin/Ethnicity'].apply(tokenizeHelper)
df['TokensDirector']=df['Director'].apply(tokenizeHelper)
df['TokensCast']=df['Cast'].apply(tokenizeHelper)
df['TokensGenre']=df['Genre'].apply(tokenizeHelper)
```

```python
df['Tokens']=df['TokensPlot'] + df['TokensTitle'] + df['TokensOrigin'] + df['TokensDirector'] + df['TokensCast'] + df['TokensGenre']
#df['Tokens']

from nltk.stem import PorterStemmer
ps = PorterStemmer()


'''def Stemmer(data):
    stem_data = []
    for item in data:
        stem_data += list(ps.stem(item))
    return stem_data
'''
df['stemmed'] = df['Tokens'].apply(lambda x: [ps.stem(word) for word in x])
#df['Stem'] = df['Tokens'].apply(Stemmer)
df['stemmed']
df_tokenized = df
```

```python
df_tokenized['Unique_Words'] = df_tokenized['stemmed'].apply(set)

def BagOfWords(uniqueWords, tokens):
    unique = tuple(uniqueWords)
    numOfWords = dict.fromkeys(unique, 0)
    for word in tokens:
        numOfWords[word] += 1
    return numOfWords

df_tokenized['Frequency'] = df.apply(lambda x: BagOfWords(x.Unique_Words, x.stemmed), axis=1)
df_tokenized['Frequency'][0]
```
